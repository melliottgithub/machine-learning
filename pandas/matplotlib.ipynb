```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
m = np.linspace(0, 10, 11)
m
```

```python
c = 3 * 10**8
E = m * c**2
```

```python
E
```

```python
plt.plot(m, E, color='red', lw=5)
plt.xlabel('Mass (kg)')
plt.ylabel('Energy (J)')
plt.title('E=mc^2')
plt.xlim(0, 10)
plt.show()
```

```python
plt.plot(m, E, color='red', lw=5)
plt.xlabel('Mass (kg)')
plt.ylabel('Energy (J)')
plt.title('E=mc^2')
plt.xlim(0, 10)
plt.yscale('log')
plt.grid(which='both', axis='y')
plt.show()
```

```python
fig= plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(m, E, color='red', lw=5)
ax.set_xlabel('Mass (kg)')
ax.set_ylabel('Energy (J)')
ax.set_title('E=mc^2')
ax.set_xlim(0, 10)
ax.set_yscale('log')
ax.grid(which='both', axis='y')
plt.show()
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(labels, july16_2007, color='red', lw=5, label='July 16, 2007')
axes.plot(labels, july16_2020, color='blue', lw=5, label='July 16, 2020')

axes.legend(loc=(1.04, 0.5))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 4))

axes[0].plot(labels, july16_2007, color='red', label='July 16, 2007')
axes[0].legend(loc=(1.04, 0.5))

axes[1].plot(labels, july16_2020, color='blue', label='July 16, 2020')
axes[1].legend(loc=(1.04, 0.5))

axes[0].set_title('Treasury Yield Curve')

```
